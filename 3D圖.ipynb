```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
x = stats.norm(0.5,0.1) #mean,std
y = stats.norm(0.5,0.1) #mean,std
z = stats.norm(0.3,0.2) #mean,std
```

```python
x_min , x_max = x.interval(0.999)
x_nor= np.linspace(x_min, x_max, 1000)
```

```python
y_min , y_max = y.interval(0.85)
y_nor= np.linspace(y_min, y_max, 1000)
```

```python
z_min , z_max = z.interval(0.999)
z_nor= np.linspace(z_min, z_max, 1000)
```

```python
xp = x.pdf(x_nor)
yp = y.pdf(y_nor)
zp = z.pdf(z_nor)
xc = x.cdf(x_nor)
yc = y.cdf(y_nor)
zc = z.cdf(z_nor)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ax = plt.subplot(111, projection='3d') 

ax.scatter(x_nor,0.1,xp)
ax.scatter(y_nor,0,yp)
ax.scatter(z_nor,3,zp)
# 
ax.set_zlabel('Z')  # 
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ax = plt.subplot(111, projection='3d') 

ax.scatter(x_nor,5,xc)
ax.scatter(y_nor,3,yc)
ax.scatter(z_nor,0,zc)
# 
ax.set_zlabel('Z')  # 
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.show()
```

```python
verts1p = list(zip(x_nor, xp))
verts2p = list(zip(y_nor, yp))
verts3p = list(zip(z_nor, zp))
vertsp = [verts1p],[verts2p],[verts3p]
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib.colors import colorConverter
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

a=[]
def cc(arg):
    return colorConverter.to_rgba(arg, alpha=0.5)

a.append(verts1p)
a.append(verts2p)
a.append(verts3p)
zs = [0.3, 0.27, 0.7]
poly = PolyCollection(a, facecolors=[cc('r'), cc('g'), cc('b')])
poly.set_alpha(0.5)                  #          
ax.add_collection3d(poly , zs=zs, zdir='y')
ax.set_xlabel('X')
ax.set_xlim3d(0, 1)
ax.set_ylabel('Y')
ax.set_ylim3d(0, 1)
ax.set_zlabel('Z')
ax.set_zlim3d(0, 4)

plt.show()
```
